```python
"""
    The purpose of this code is to generate the number of folds
    required for a paper's thickness to exceed the height of
    Mount Everest.
"""

# code to get the number of folds of a paper if its thickness exceed the height of Mount Everest

thickness = 1*(10**-3) # initial_thickness
no_of_folds = 0 # used to store or hold the number of folds

# code to get the thickness value

while thickness < 8848:
    no_of_folds += 1
    thickness *= 2

print("The number of folds required for a 1mm thick paper to exceed the height of Mount Everest is: ", no_of_folds)
```

```python
"""
    For a reservoir whose water volume decreases exponentially,
    this code tries to solve the time taken for the water in
    the reservoir to diminish to half its original volume
"""

import math

# To get the time, I have deduced the forumla as t = -ln(vratio)/alne, where vratio is
# the volume ratio of the reservoir which is 1/2 or 0.5

vratio = 0.5 # ratio between final and inital volume of the reservoir
a = 0.1 # some random constant of the exponential process

t = -math.log(vratio)/(a*math.log(math.e)) # time for the volume to diminish exponentially

print("The time required for the reservoir to diminish to less than half its volume is: ", t)
```

```python
"""
    Calculate the compound interest in 3, 4, and 5 years of a princpal amount of 
    money worth $100 at an interest rate of 1%.
"""

principal = 100 # The principal deposit
interest = 0.05 # The interest rate
years = 0 #initalizing the number of years

# enclosing the compound interest formula in a function
def final_money(principal, interest, years):
    """
    This function uses the popular simple interest formula
    to calculate the final money in a certain number of years
    after a principal amount has been deposited at the bank.
    """
    final_money = principal*(1+interest)**years # This represents the final money in the account in n years
    return final_money

money1 = round(final_money(principal, interest, 1)) # final money in 1 year
money2 = round(final_money(principal, interest, 2)) # final money in 2 years
money3 = round(final_money(principal, interest, 3)) # final money in 3 years
money4 = round(final_money(principal, interest, 4)) # final money in 4 years
money5 = round(final_money(principal, interest, 5)) # final money in 5 years

print("The final money reflected in the acount in 1 year is:  ", money1, "dollars")
print("The final money reflected in the acount in 2 years is: ", money2, "dollars")
print("The final money reflected in the acount in 3 years is: ", money3, "dollars")
print("The final money reflected in the acount in 4 years is: ", money4, "dollars")
print("The final money reflected in the acount in 5 years is: ", money5, "dollars")
```

```python
"""
    This code calculates the monthly payment to pay
    off a debt in 1, 2 and 3 years if the initial loan
    is $20,000 and the monthly interest rate is 1%
"""

principal = 20000 # initial loan amount
period_interest = 0.01 # monthly interest rate

# function to define the debt to be paid in a certain number of years
def month_pay(principal, period_interest, periods):
    """
    This function describes the monthly payment required to pay up a given loan.
    It is based on the formula for amortized loan.
    
    principal: initial loan
    period_interest: interest rate
    periods: number of months, years, or days
    """
    final_pay = (principal*period_interest)/(1-(1+period_interest)**(-periods)) # the amount to be paid monthly to pay off the debt
    return final_pay

monthly1 = round(month_pay(principal, period_interest, 12))
monthly2 = round(month_pay(principal, period_interest, 24))
monthly3 = round(month_pay(principal, period_interest, 36))



print("The monthly payment required to pay up the debt in 1 year is:  ", monthly1, "dollars")
print("The monthly payment required to pay ub the debt in 2 years is: ", monthly3, "dollars")
print("The monthly payment required to pay ub the debt in 3 years is: ", monthly2, "dollars")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

"""
    Code to get number of days it will take to 
    repay initial investments based on cumulated profits.
"""

initial_investment = 100000 # initial amount invested
customer = 100 # initial number of customers

growth = 0.01 # rate at which customers grow
profit_customer = 10 # profit contributed per customer

profit = profit_customer*customer # initial profit made off 100 customers

day = 1 # day 1 of the business

cum_profit = np.array([profit]) # array of cumulated profit values

while profit<initial_investment:
    extra_customer = growth*customer # extra customers added after each day
    customer += extra_customer
    profit += profit_customer*customer 
    profit_arr = np.array([profit])  # cumulated profit to append to a numpy array
    cum_profit = np.append(cum_profit,profit_arr) # appending our cumulated profit into an array
    day +=1

    
print("Number of days required to cover initial investment is: ", day)

ndays = np.arange(1,day+1) # number of days
cpft = cum_profit # cumulated profit

# create a figure and axes
fig, ax = plt.subplots(figsize=(10,6))

# setting graph's title
ax.set_title("Graph plot of Cumulated Profit versus days")

# labelling of x and y axis
ax.set_ylabel("Cumulated Profits")
ax.set_xlabel("Number of Days")

ax.set_xlim(1,day+3) # setting the limits of the x axis
# plotting of graph
ax.plot(ndays, cpft)

ax.hlines(y=100000, xmin=0, xmax=day+3, colors='r', linestyles='--',lw=2, label="Initial Investment")
ax.vlines(x=69.66, ymin=0, ymax=cum_profit[-1], colors='b', linestyles='--',lw=2, label="Breakeven Day")

# Add labels to the plot
style = dict(size=10, color='black')
ax.text(day+5, 99000, "Initial Investment", **style)

ax.annotate('Breakeven Day', xy=(69.66, 0), xytext=(80, 30000),
 arrowprops=dict(facecolor='black', shrink=0.05))

plt.scatter(69.66,0, color='g')

plt.legend(bbox_to_anchor=(1.04,0.5), loc="center left", borderaxespad=0)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

"""
    This code aims to analyze the number of Ebola cases and deaths caused
    by Ebola by plotting it out, and estimating the data against certain thresholds.
"""

data = pd.read_excel('ebola_download.xls', usecols="A:C") # reading of our ebola data
data['Date'] = pd.to_datetime(data['Date']) #converting our date column to a datetime object

data = data.set_index('Date') # setting date as our index
data = data.reindex(pd.date_range('2014-03-22','2014-11-12',freq='D')) # reindexing date to have new values that were missing
data.index.name = 'Date' # renaming our index to Date
data = data.reset_index() # pushes Date into the main dataframe

data["Cases"] = data["Cases"].interpolate(method = "linear") #interpolation to estimate ebola cases
data["Death"] = data["Death"].interpolate(method = "linear") # interpolation to estimate missing ebola deaths

fig, ax = plt.subplots(figsize=(10,8))

# setting graph's title
ax.set_title("Graph plot of Date against Number of Cases and Deaths")

# labelling of x and y axis
ax.set_ylabel("Number of Ebola Cases or Death")
ax.set_xlabel("Date")


# plotting of graph of Ebola Cases and Deaths
ax.plot(data['Date'],data['Cases'], label='Ebola Cases')
ax.plot(data['Date'],data['Death'], label='Ebola Deaths')

# Add labels to the plot
# These labels represent test cases where Ebola cases and deaths exceed
# 100, 500, 1000, 2000 and 5000 respectively.

cases =[100, 500, 1000, 2000, 5000]
color = ['g', 'b', 'r', 'k', 'y'] # Colors to be used to mark our test cases on the graph

for i,j in zip(cases, color):
    if j=='g':
        label='Number of Ebola Cases and Deaths > 100'
    elif j=='b':
        label = 'Number of Ebola Cases and Deaths > 500'
    elif j=='r':
        label='Number of Ebola Cases and Deaths > 1000'
    elif j=='k':
        label = 'Number of Ebola Cases and Deaths > 2000'
    else:
        label = 'Number of Ebola Cases and Deaths > 5000'
        
    k=data.loc[data['Cases']>i] # filtering datframe that fulfills case i
    k_index=k.iloc[0] # getting a Pandas series that fulfills i
    ax.scatter(k_index['Date'], k_index['Cases'], color=j, label=label) # plotting test case i for Ebola Cases
    l=data.loc[data['Death']>i]
    l_index=l.iloc[0]
    ax.scatter(l_index['Date'], l_index['Death'], color=j) # plotting test case i for Ebola Deaths
    
ax.legend()

# ebola cases
for i in cases:
    k = data.loc[data['Cases']>i]
    k_index = k.iloc[0]
    print("Ebola Cases exceeded ", i, "at estimated date: ", k_index['Date'].date())
    
print("\n") # To make output cleaner
    
# ebola deaths
for i in cases:
    k = data.loc[data['Death']>i]
    k_index = k.iloc[0]
    print("Ebola Deaths exceeded ", i, "at estimated date: ", k_index['Date'].date())
    print(k_index['Date'].date())
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.dates as mdates

"""
    This code aims to plot the normalized time series of two stocks
    in order to give room for comparisons
"""

spy = pd.read_csv("SPY.csv", usecols=["Date","Adj Close"]) # Reading of the stock SPY dataset
tlt = pd.read_csv("TLT.csv", usecols=["Date","Adj Close"]) # Reading of the stock TLT datset

# definition of function to normalize datasets SPY and TLT
def normalize(a):
    min = a["Adj Close"].min() # finding the minimum Adj Close price in the dataset
    scaled_a = a.iloc[:,1:].div(min) # dividing the Adj Close price with the minimum value observed 
    normalize_a = scaled_a.iloc[0:]*100 # multiplying all the elements of the dataset by 100 to normalize it
    return normalize_a

norm_spy = normalize(spy) # normalizing the SPY dataset
norm_tlt = normalize(tlt) # normalizin the TLT dataset
date_spy = spy["Date"] # Obtaining the date column for plotting purposes

fig, ax = plt.subplots(figsize=(20,8))

# setting graph's title
ax.set_title("Graph plot of Closing Prices against Date")

# labelling of x and y axis
ax.set_ylabel("Prices")
ax.set_xlabel("Date")

# plotting of graph
ax.plot(date_spy,norm_spy["Adj Close"], label='SPY') # plotting the normalized prices for the SPY dataset
ax.plot(date_spy,norm_tlt["Adj Close"], label='TLT') # plotting the normalized prices for the TLT dataset

# fixing the clustering of date intervals on the x axis
index = pd.date_range(start = date_spy[0], end=date_spy[len(date_spy)-1], freq='M') 
index = [pd.to_datetime(date, format='%Y-%m-%d').date() for date in index]
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=1))

ax.legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.dates as mdates

"""
    This code aims to compute the average, minimum and maximum daily returns for the 
    two ETF stocks.
"""

spy = pd.read_csv("SPY.csv", usecols=["Date","Adj Close"]) # Reading of the stock SPY dataset
tlt = pd.read_csv("TLT.csv", usecols=["Date","Adj Close"]) # Reading of the stock TLT datset

# definition of function to normalize datasets SPY and TLT
def normalize(a):
    min = a["Adj Close"].min() # finding the minimum Adj Close price in the dataset
    scaled_a = a.iloc[:,1:].div(min) # dividing the Adj Close price with the minimum value observed 
    normalize_a = scaled_a.iloc[0:]*100 # multiplying all the elements of the dataset by 100 to normalize it
    return normalize_a

norm_spy = normalize(spy) # normalizing the SPY dataset
norm_tlt = normalize(tlt) # normalizin the TLT dataset
date_spy = spy["Date"] # Obtaining the date column for plotting purposes


# definition of the daily return function
def dreturn(a):
    result = [] # an empty list to store our daily returns
    for i in range(len(a)):
        b = i-1
        if b<0:
            dreturn = float("Nan")
            result.append(dreturn)
        else:
            dreturn = (a["Adj Close"].iloc[i]/a["Adj Close"].iloc[b])-1 # implementation of daily return formula
            result.append(dreturn) # appending each daily return to result varaible
        final_result= pd.DataFrame(result) # making a dataframe from our list of results
        final_result.columns=['Daily Returns'] # our dataframe storing values of our daily returns
    return final_result
   

norm_spy = normalize(spy) # normalizing the ETF SPY
norm_tlt = normalize(tlt) # normalizing the ETF TLT

retSpy=dreturn(norm_spy) # daily return of the ETF SPY
retTlt=dreturn(norm_tlt) # daily return of the ETF TLT

minSpy = retSpy.min()[0] # minimum daily return of the ETF SPY
maxSpy = retSpy.max()[0] # maximum daily return of the ETF SPY
avgSpy = retSpy.mean()[0] # average daily return of the ETF SPY

minTlt = retTlt.min()[0] # minimum daily return of the ETF TLT
maxTlt = retTlt.max()[0] # maximum daily return of the ETF TLT
avgTlt = retTlt.mean()[0] # average daily return of the ETF TLT


print("The minimum value for the ETF STOCK, SPY is: ", minSpy)
print("The maximum value for the ETF STOCK, SPY is: ", maxSpy)
print("The average value for the ETF STOCK, SPY is: ", avgSpy)

print("The minimum value for the ETF STOCK, TLT is: ", minTlt)
print("The maximum value for the ETF STOCK, TLT is: ", maxTlt)
print("The average value for the ETF STOCK, TLT is: ", avgTlt)

```
